```python
# plot the normal theoretical quantiles against the sample quantiles of a N(3,4) and a Exp(4) distribution
from scipy.stats import norm, expon
import numpy as np
import matplotlib.pyplot as plt

# generate 1000 random numbers from a N(3,4) distribution
x = norm.rvs(loc=3, scale=2, size=99)
# generate 1000 random numbers from a Exp(4) distribution
y = expon.rvs(scale=1/4, size=99)

# sort the numbers
x.sort()
y.sort()

# generate the theoretical quantiles
q_theor = norm.ppf(np.arange(0.01, 1, 0.01), loc=0, scale=1)

# plot the theoretical quantiles against the sample quantiles
# plt.plot(q_theor, x, 'o', color='red', label='N(3,4)')

# normal quantiles vs sample quantiles of Exp(4)
plt.plot(q_theor, x, 'o', color='blue', label='Exp(4)')
```
